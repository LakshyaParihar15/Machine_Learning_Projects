```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset = pd.read_csv('data.csv')
dataset
```

```python
dataset.info()
```

```python
Dataset = dataset.drop('date', axis=1)
```

```python
print(Dataset)
```

```python
dataset.describe()
```

```python
Y = dataset.iloc[:, 1].values
print(Y)
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.impute import SimpleImputer
import numpy as np

# Select categorical and numerical columns
categorical_columns = ['street', 'city', 'statezip', 'country']
numerical_columns = ['bedrooms', 'bathrooms', 'sqft_living', 'sqft_lot', 'floors', 
                     'waterfront', 'view', 'condition', 'sqft_above', 'sqft_basement', 'yr_built', 'yr_renovated']

# Create a ColumnTransformer to handle both categorical and numerical columns
ct = ColumnTransformer(
    transformers=[
        ('num', SimpleImputer(strategy='mean'), numerical_columns),  # Impute missing numerical data with the mean
        ('cat', OneHotEncoder(sparse_output=False, handle_unknown='ignore'), categorical_columns)  # One-hot encode categorical data
    ],
    remainder='passthrough'  # Keep the remaining columns (e.g., 'price')
)

# Apply the preprocessing
X_preprocessed = ct.fit_transform(Dataset)

# Display the preprocessed data
X_preprocessed[:5]
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X_preprocessed, Y, test_size=0.2, random_state=2)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train= sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(x_train, y_train)
```

```python
y_pred = regressor.predict(x_test)
print(y_pred)
```

```python
from sklearn.metrics import r2_score

r2 = r2_score(y_test, y_pred)
print("R² Score:", r2)
```
